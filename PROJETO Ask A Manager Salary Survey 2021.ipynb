```python
values_to_replace = ['USA', 'US','U..S.','Us','U. S','Usa','United States of America','Usat','Us','Y','usa','U.S.','USA ','U. S ','Usa', 'united states',
                     'UNITED STATES','bonus based on meeting yearly goals set w/ my supervisor','United State of America','United Stated','United Statees',
                     'America','California','Currently finance','ISA','The United States','Remote','Uniter Statez','United States (I work from home and my clients are all over the US/Canada/PR',
                     'United Sates','I was brought in on this salary to help with the EHR and very quickly was promoted to current position but compensation was not altered. ',
                     'U.S.A','U.A.','IS','U.S>','US of A','United Stateds','United Statea','U.S.A.','United States of America',
                     'United Stattes','United Statues','UnitedStates','United Statws','United y','United Statws','United Sttes']

# Substituir os valores na coluna 'Country' por 'United States'
df_EstadosUnidos.loc[df_EstadosUnidos['Country'].isin(values_to_replace), 'Country'] = 'United States'
```

```python
df_EstadosUnidos = df_EstadosUnidos[df_EstadosUnidos['Job title'] != 'bum']
```

```python
df_EstadosUnidos = df_EstadosUnidos[df_EstadosUnidos['Country'] == 'United States']
```

```python
# Especifica o nome do arquivo e o caminho onde deseja salvar o arquivo Excel

file_name = 'Dados_sal√°rios_USA.xlsx'

# Salva o DataFrame como um arquivo Excel
df_EstadosUnidos.to_excel(file_name, index=False)
```

```python
# Selecionar as 10 ind√∫strias que pagam mais
top_10_industries_salary = df_EstadosUnidos.groupby('Industry')['Annual Salary'].median().sort_values(ascending=False).head(10)
print(top_10_industries_salary)

fig = px.bar(top_10_industries_salary, 
             x=top_10_industries_salary.values, 
             y=top_10_industries_salary.index, 
             orientation='h', 
             title='Top 10 Industrias por m√©dia de sal√°rio (Estados Unidos)', 
             labels={'x': 'Median Salary', 'y': 'Industry'}, 
             color=top_10_industries_salary.values, 
             color_continuous_scale='Viridis')

fig.update_layout(coloraxis_showscale=False)
fig.show()
```
